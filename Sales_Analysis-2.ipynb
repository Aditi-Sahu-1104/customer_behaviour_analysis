```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", '_')
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

## creating new column

```python
labels = ['Young-Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df.head()
```

## purchase_frequency_column

```python
df['frequency_of_purchases'].unique()
```

```python
## mapping data

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Annually' : 365,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Every 3 Months' : 90
 }

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
query = '''select item_purchased,
round(avg(review_rating), 2) AS "average_rating"
from customer_behaviour
group by item_purchased
order by avg(review_rating) desc
limit 5'''
pd.read_sql_query(query, conn)
```

```python
query = '''select shipping_type, round(avg(purchase_amount),2)
from customer_behaviour
where shipping_type in ('Standard', 'Express')
group by shipping_type
'''
pd.read_sql_query(query, conn)
```

```python
query = '''select subscription_status,
count(customer_id) AS "total_customer", 
round(avg(purchase_amount), 2) As "average_spent", 
sum(purchase_amount) AS "total_revenue"
from customer_behaviour
group by subscription_status'''
pd.read_sql_query(query, conn)
```

```python
query= '''select item_purchased,
round(100 * sum(case when discount_applied = "Yes" then 1 else 0 end)/count(*),2) AS "discount_rate"
from customer_behaviour
group by item_purchased
order by discount_rate desc
limit 5
'''

pd.read_sql_query(query, conn)
```

```python
query = '''
with customer_type as(
select customer_id, previous_purchases,
case
when previous_purchases = 1 then "New"
when previous_purchases between 2 and 10 then "Returning"
else "Loyal"
end as customer_segment
from customer_behaviour
)
select customer_segment, count(*) AS "number_of_customers"
from customer_type
group by customer_segment'''

pd.read_sql_query(query, conn)
```

```python
query = '''
with item_count as(
select category,
item_purchased,
count(customer_id) as "total_orders",
row_number() over (partition by category order by count(customer_id) desc) as "item_rank"
from customer_behaviour
group by category, item_purchased
)
select item_rank, category, item_purchased, total_orders
from item_count
where item_rank <=3'''

pd.read_sql_query(query, conn)
```

```python
query = '''
select subscription_status,
count(customer_id) as repeat_buyer
from customer_behaviour
where previous_purchases > 5
group by subscription_status'''

pd.read_sql_query(query, conn)
```

```python
query = '''
select age_group,
sum(purchase_amount) as total_revenue
from customer_behaviour
group by age_group
order by total_revenue desc'''

pd.read_sql_query(query, conn)
```
